# A Notebook of Linear Regression Experiments


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# A function to create a scatter plot of the data with a regression line
def create_scatter_plot(x, y):
    """
    A function to create a scatter plot of the data with a regression line.
    The function creates a figure, a scatter plot, a regression line, and adds title and labels.
    Args:
        x: The x-values of the data
        y: The y-values of the data
    Returns:
        None
    """
    # Validate the data
    if len(x) == 0 or len(y) == 0:
        raise ValueError("Empty input data")

    # Create a figure
    plt.figure(figsize=(10, 6))

    # Create a scatter plot with large blue dots
    plt.scatter(x, y, color='blue', s=100, alpha=0.5, label='Data points')

    # Prepare data for linear regression by reshaping it for sklearn
    x_array = np.array(x).reshape(-1, 1)

    # Create and fit the linear regression model
    model = LinearRegression()
    model.fit(x_array, y)

    # Generate predictions
    y_pred = model.predict(x_array)

    # Plot the regression line
    plt.plot(x, y_pred, color='black', linewidth=2, label='Regression line')

    # Add title and labels
    plt.title('Linear Regression Model')
    plt.xlabel('X values')
    plt.ylabel('Y values')
    plt.legend()

    # Show the plot
    plt.show()

# Your data
x = [3, 5, 6, 7, 9, 10, 11]
y = [4, 6, 7, 8, 10, 11, 9]

def main():
    create_scatter_plot(x, y)

main()











```
